```python
dataaa = pd.read_csv('iris.csv')
```

#To Standardize the Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
features = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
```

```python
x = dataaa.loc[:, features].values


#extracting the features
```

```python
x[0:10]
```

```python
x_standardized = StandardScaler().fit_transform(x)
```

```python
x_standardized[0:10]
```

#Covariance Matrix computation

```python
#where T is Transpose

import numpy as np
cov_data = np.corrcoef(x.T)
cov_data
```

```python
import numpy as np
cov_data = np.corrcoef(x_standardized.T)
cov_data
```

```python
import matplotlib.pyplot as plt
img = plt.matshow(cov_data, cmap=plt.cm.rainbow)
plt.colorbar(img, ticks = [-1, 0, 1], fraction=0.045)
for x in range(cov_data.shape[0]):
    for y in range(cov_data.shape[1]):
        plt.text(x, y, "%0.2f" % cov_data[x,y], size=12, color='black', ha="center", va="center")
        
plt.show()

#[0] is the row while [1] is the column
#"%0.2f" specifies the number of decimal point. see below for example
```

```python
import matplotlib.pyplot as plt
img = plt.matshow(cov_data, cmap=plt.cm.rainbow)
plt.colorbar(img, ticks = [-1, 0, 1], fraction=0.045)
for x in range(4):
    for y in range(4):
        plt.text(x, y, "%0.2f" % cov_data[x,y], size=12, color='black', ha="center", va="center")
```

#PCA Projection to 2D

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components=pca.fit_transform(x_standardized)
principalDF=pd.DataFrame(data=principal_components, columns= ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDF, dataaa[['variety']]], axis = 1)
finalDf
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principal_components=pca.fit_transform(x_standardized)
principalDF=pd.DataFrame(data=principal_components, columns= ['principal component 1', 'principal component 2', 'principal componet 3'])
finalDf = pd.concat([principalDF, dataaa[['variety']]], axis = 1)
finalDf
```

#Visualize PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components=pca.fit_transform(x_standardized)
principalDF=pd.DataFrame(data=principal_components, columns= ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDF, dataaa[['variety']]], axis = 1)
finalDf
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Setosa', 'Versicolor', 'Virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['variety'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
